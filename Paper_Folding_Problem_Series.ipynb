```python
# CODE > PROBLEM 5

THICKNESS = 0.00008

folded_thickness = THICKNESS

# Lista para almacenar el grosor después de cada pliegue
thickness_list = [folded_thickness]

# Dobla el papel 43 veces
for _ in range(43):
    folded_thickness *= 2
    thickness_list.append(folded_thickness)

print("Thickness before  43 Folds: {} meters".format(folded_thickness))

#CONVERT METERS TO KILOMETERS, AND DISPLAY WITH TWO DECIMAL PLACES
print("THICKNESS ON KILOMETERS: {:.2f} kilometers".format(folded_thickness / 1000))

# Verifica la cantidad de valores en la lista
print("Cantidad de valores en la lista:", len(thickness_list))
```

```python
# CODE > PROBLEM 6

import matplotlib.pyplot as plt

THICKNESS = 0.00008
folded_thickness = THICKNESS

# Lista para almacenar el grosor después de cada pliegue
thickness_list = [folded_thickness]

# Dobla el papel 43 veces
for _ in range(43):
    folded_thickness *= 2
    thickness_list.append(folded_thickness)

# Crear una lista de pliegues (0 a 43)
folds = list(range(len(thickness_list)))

# Visualizar el crecimiento exponencial
plt.figure(figsize=(10, 6))
plt.plot(folds, thickness_list, marker='o')
plt.title('Crecimiento Exponencial del Grosor del Papel')
plt.xlabel('Número de Pliegues')
plt.ylabel('Grosor del Papel (metros)')
plt.yscale('log')  # Usar escala logarítmica para visualizar mejor el crecimiento
plt.grid(True)
plt.xticks(folds)  # Mostrar todos los números de pliegues en el eje X
plt.show()

# Verifica la cantidad de valores en la lista
print("Cantidad de valores en la lista:", len(thickness_list))

# THE CHANGE OF THICKNESS ARE DESCRIBING EXPONENTIAL RICING ( y= X^2)
```
